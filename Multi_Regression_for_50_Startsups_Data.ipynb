### 5.2 Model Validation:
    Three Techniques: 
    1. Collinearity Check 
    2. Residual Analysis
    3. Homoscedasticity

### 1) Collinearity Problem Check
Calculate VIF = 1/(1-Rsquare) for all independent variables

```python
rsq_rds=smf.ols("RDS~ADMIN+MKTS",data=startups_details_1).fit().rsquared
vif_rds=1/(1-rsq_rds)

rsq_admin=smf.ols("ADMIN~RDS+MKTS",data=startups_details_1).fit().rsquared
vif_admin=1/(1-rsq_admin)

rsq_mkts=smf.ols("MKTS~RDS+ADMIN",data=startups_details_1).fit().rsquared
vif_mkts=1/(1-rsq_mkts)


# Putting the values in Dataframe format:

df_1={'Variables':['RDS','ADMIN','MKTS'],'VIF':[vif_rds,vif_admin,vif_mkts]}
Vif_df_1=pd.DataFrame(df_1)
Vif_df_1
```

```python
# From above table, None variable have VIF>20, So No Collinearity. Consider all varaibles in Regression equation.
```

### 5.3 Model Deletion Diagnostics (checking Outliers or Influencers)
Two Techniques : 1. Cook's Distance and 2. Leverage value

```python
# 1) Cook's Distance: If Cook's distance is found > 1, then it is an Outlier.
# Get influencers using cook's distance:

(c,_)=model.get_influence().cooks_distance
c
```

```python
# Plotting influencers using stem plot:

fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(startups_details_1)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
import warnings
warnings.filterwarnings('ignore')
```

```python
# Index and Value of Influencer, Where C > 0.5
np.argmax(c) , np.max(c)
```

```python
# 2) Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers

influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = No.of features/columns & n = No. of Datapoints

k=startups_details_1.shape[1]
n=startups_details_1.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

## Improving the model:

```python
# Discard data points which are influencers and re-assign row number (reset_index(drop=True))

startups_details_2=startups_details_1.drop(startups_details_1.index[[49]],axis=0).reset_index(drop=True)
startups_details_2
```

## Model Deletion Diagnostics and Final Model:

```python
while np.max(c)>0.5 :
    model=smf.ols("Profit~RDS+ADMIN+MKTS",data=startups_details_2).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    startups_details_2=startups_details_2.drop(startups_details_2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    startups_details_2
else:
    final_model=smf.ols("Profit~RDS+ADMIN+MKTS",data=startups_details_2).fit()
    final_model.rsquared , final_model.aic
    print("Model accuracy is improved to",final_model.rsquared)
```
